x,y,z merupakan panjang, lebar, dan kedalaman sehingga tidak memungkinkan nilai min nya itu 0

```python
x = (dia.x == 0).sum()
y = (dia.y == 0).sum()
z = (dia.z == 0).sum()

print("Nilai 0 di kolom x ada di: ", x)
print("Nilai 0 di kolom y ada di: ", y)
print("Nilai 0 di kolom z ada di: ", z)
```

```python
dia.loc[(dia['z']==0)]
```

seluruh data bernilai 0 pada dimensi x dan y juga memiliki nilai 0 pada dimensi z

```python
dia
```

kita drop invalid data z=0 karena sudah mewakili nilai x dan y yang 0

```python
dia = dia.loc[(dia[['z']] != 0).all(axis=1)]

# cek ukuran
dia.shape
```

```python
# Select only numeric columns
numeric_cols = dia.select_dtypes(include=[np.number]).columns

Q1 = dia[numeric_cols].quantile(0.25)
Q3 = dia[numeric_cols].quantile(0.75)
IQR = Q3 - Q1  # IQR is interquartile range.

dia = dia[~((dia[numeric_cols] < (Q1 - 1.5 * IQR)) | (dia[numeric_cols] > (Q3 + 1.5 * IQR))).any(axis=1)]

# cek ukuran
dia.shape
```

```python
numerical_features = ['price', 'carat', 'depth', 'table', 'x', 'y', 'z']
category_features = ['cut', 'color', 'clarity']
```

```python
feature = category_features[0]
count = dia[feature].value_counts()
percent = 100*dia[feature].value_counts(normalize=True)
df = pd.DataFrame({'Jumlah Sampel': count, 'persentase': percent.round(1)})
print(df)
count.plot(kind='bar', title=feature);
```
